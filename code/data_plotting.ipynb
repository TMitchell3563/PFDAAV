# Class to plot data

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class DataPlotter:
    def __init__(self, filepath):
        try:
            self.df = pd.read_csv(filepath)
            self.df.set_index(self.df.columns[0], drop = True, inplace = True)
        except Exception as e:
            print('File could not be read')
            print(f'Error: {e}')

    def plot_bar_chart(self, col):
        # Get the value counts for the column
        counts = self.df[col].value_counts().sort_index()

        # Plotting
        plt.figure(figsize=(8, 6))
        counts.plot(kind='bar')
        plt.title('Number of Occurrences')
        plt.xlabel('Values')
        plt.ylabel('Occurrences')
        plt.xticks(rotation=0)  # Rotate x labels if needed
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
        
    def plot_pie_chart(self, col):
        # Get the value counts for the column
        counts = self.df[col].value_counts().sort_index()

        max = counts.idxmax()
        if counts[max]) / sum(counts) > 0.75:
            print(f'Plotting exploded pie chart due to dominance of category {max}')
            # Highlight the dominant category
            explode = [0.1 if x == max else 0 for x in counts.index]
            # Plot the pie chart            
            plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode)
        else:
            # Plotting
            plt.figure(figsize=(8, 6))
            plt.pie(counts, labels = counts.index, autopct='%1.1f%%', startangle=140)
        plt.title('Distribution of Occurrences')
        plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()  

```
